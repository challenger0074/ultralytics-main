```python
video=cv2.VideoCapture('./yolotest.mp4')
num=0
save_step=30
while True:
    ret,frame=video.read()
    if not ret:
        break
    if num%save_step==0:
        cv2.imwrite(f'./demoImg/yolotest_{num}.jpg',frame)
    num+=1
```
